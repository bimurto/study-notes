```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# A simple line
# Prepare the data
x = np.linspace(0, 10, 100)
y = np.linspace(0, 50, 100)

# Plot the data
plt.plot(x, y, label='linear')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

```python
# Adding some extra features
# Prepare the data
x = np.linspace(0, 10, 100)
y = np.linspace(0, 50, 100)

# Plot the data
plt.plot(x, y, c='r', linestyle='--',label='linear', linewidth=2)

# Add a legend
plt.legend()
plt.title('Basic Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.xlim(0,12)
plt.ylim(0,60)
plt.grid(True)

# Show the plot
plt.show()
```

## Plots could be customised at three levels:
### Colours
* b – blue
* c – cyan
* g – green
* k – black
* m – magenta
* r – red
* w – white
* y – yellow
* Can use Hexadecimal, RGB formats
### Line Styles
* ‘-‘ : solid line
* ‘- -‘: dotted line
* ‘- .’: dash-dot line
* ‘:’ – dotted line
### Marker Styles
* .  – point marker
* ,  – Pixel marker
* v – Triangle down marker
* ^ – Triangle up marker
* < – Triangle left marker
* \> – Triangle right marker
* 1 – Tripod down marker
* 2 – Tripod up marker
* 3 – Tripod left marker
* 4 – Tripod right marker
* s – Square marker
* p – Pentagon marker
* \* – Star marker
### Other configurations
* color or c
* linestyle
* linewidth
* marker
* markeredgewidth
* markeredgecolor
* markerfacecolor
* markersize

```python
# Multiple plots on one figure
x= np.arange(0,1500,100)
plt.plot(x, np.sin(x),label='sin function x')
plt.plot(x, np.cos(x),label='cos functon x')
plt.legend(loc='upper right')
plt.show()
```

```python
# Create subplots
fig, axs = plt.subplots(2)
#create data
x=np.linspace(0,100,10)

# assign the data to the plot using axs
axs[0].plot(x, np.sin(x**2))
axs[1].plot(x, np.cos(x**2))

# add a title to the subplot figure
fig.suptitle('Vertically stacked subplots')
plt.show()
```

```python
# Change plot figure info
# The object figure is a container for showing the plots and is instantiated by calling figure() function
fig = plt.figure(figsize=(4,3))

# create a sample data
X = np.array([1,2,3,4,5,6,8,9,10])
Y = X**2

# plot the figure
plt.plot(X,Y)
plt.show()

fig = plt.figure(figsize=(10,3),facecolor='y',edgecolor='r',linewidth=5)
plt.plot(X,Y)
plt.show()
```

```python
# bar chart
subject = ['maths','english','science','social','computer']
marks =[70,80,50,30,78]
plt.bar(subject,marks)
plt.show()
```

```python
# Pie chart
Tickets_Closed = [10, 20, 8, 35, 30, 25]
Agents = ['Raj', 'Ramesh', 'Krishna', 'Arun', 'Virag', 'Mahesh']

# create pie chart
plt.pie(Tickets_Closed, labels = Agents)
plt.show()
```

```python
# Scatter plot
x = np.random.randn(1000)
y = np.random.randn(1000)
colors = 100 * np.random.randn(1000)
plt.scatter(x,y,c=colors)
plt.show()
```

```python
# histogram
data = np.random.randn(10000)
plt.hist(data, bins=300)
plt.show()
```
